```python
import pandas as pd 
import pickle
import fonctions_utiles as fu 
import matplotlib.pyplot as plt
departement=input("Entrez le num√©ro du d√©partement (ex : 01, 2A, 36...) : ")
selected_path=f"/Users/louisedeferran/Documents/Louise/Mines/effynersys/reseau_chaleur/DHN_Potential_France_EFFINERSYS/National_Scale/Storage_results/dep_{departement}/selected_communes.csv"
selected_communes = pd.read_csv(selected_path)
liste_communes = selected_communes['commune'].tolist()

fig, axes = plt.subplots(2, 5, figsize=(20, 8))  # 10 graphes
axes = axes.flatten()

for ax, commune in zip(axes, liste_communes[:11]):
    print(f"Traitement de la commune {commune}...")
    pickle_path = f"/Users/louisedeferran/Documents/Louise/Mines/effynersys/reseau_chaleur/DHN_Potential_France_EFFINERSYS/National_Scale/Storage_results/dep_{departement}/communes/{commune}.pkl"
    with open(pickle_path, "rb") as f:
        result = pickle.load(f)
    plt.sca(ax)
    fu.tracer_graphe_optimal_zoom(
        commune=commune,
        result=result
    )
    ax.set_title(f"Commune {commune}")
plt.tight_layout()
plt.show()

```
